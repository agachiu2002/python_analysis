```python
import pandas as pd

# 建立 DataFrame (Let's create the DataFrame)
data = {'姓名': ['小明', '小美', '大雄', '靜香', '胖虎'],
        '年齡': [25, 30, 25, 35, 40],
        '城市': ['台北', '高雄', '台中', '台北', '高雄']}
df = pd.DataFrame(data)

# 建立布林條件 (Create a Boolean condition)
condition = df['城市'] == '台北'

# 使用布林索引篩選資料 (Filter the DataFrame using Boolean indexing)
taipei_data = df[condition]

# 顯示篩選結果 (Display the filtered data)
print(taipei_data)
```
