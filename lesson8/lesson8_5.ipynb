```python
import numpy as np
import pandas as pd

scores = np.random.randint(50, 101, size=(50, 5))
df = pd.DataFrame(scores,
             columns=['國文', '英文', '數學', '地理', '歷史'],
             index=range(1,51)
             )
df.index.name = '學號'
df.columns.name = '科目'
df
```

```python
#排序
sum_value = df.sum(axis=1)
mean_value =df.mean(axis=1)
```

```python
df['總分']= sum_value
df['平均']= mean_value
```

```python
df1=df.sort_values(by='總分',ascending=False)
df1
```

```python
#排名
rand_value = df1['總分'].rank(ascending=False,method='min')
df1['排名'] =rand_value
df1
```

```python
def max_score(row) ->int:
    #return row.max()  #抓這個人的最高分
    #return row.std() #抓標準差
    return pd.Series ([row.min(),row.max(), row.std()],index=['最低分','最高分','標準差'])
    

df1[['國文','英文','數學','地理','歷史']].apply(max_score,axis=1)
```

```python
def max_score(row):
    #return row.max()  #抓這個人的最高分
    #return row.std() #抓標準差
    less_60_count = len(row[row<60])
    return pd.Series ([row.min(),row.max(), row.std(),less_60_count],index=['最低分','最高分','標準差','不及格的數量'])
    
df2=df1[['國文','英文','數學','地理','歷史']].apply(max_score,axis=1)

pd.concat([df1,df2],axis=1)
```
