style.format() 方法的常用格式化字串

以下是一些常用的格式化字串及其說明：

{:.2f}： 將浮點數顯示為兩位小數。
{:.0f}： 將浮點數顯示為整數（無小數）。
{:,}： 添加千位分隔符。
{:.2%}： 將數值顯示為百分比，並保留兩位小數。
${:.2f}： 添加美元符號，並保留兩位小數。
{:.2e}: 科學記號表示法，並保留兩位小數。
{:+.2f}: 顯示正負號，並保留兩位小數。
{:0>4d}: 顯示整數，不足四位數，前面補0。
{:%Y/%m/%d}： 將日期顯示為 YYYY/MM/DD 格式。
{:%Y-%m-%d %H:%M:%S}: 將日期時間顯示為 YYYY-MM-DD HH:MM:SS 格式。
lambda x: f'{x:.2f}': 使用lambda函數，將浮點數顯示為兩位小數。

```python
def rain_condition(v):
    if v < 1.75:
        return "乾燥"
    elif v < 2.75:
        return "有雨"
    else:
        return "大雨"

def change_index(t):
   return t.strftime('%Y年%m月%d日')

weather_df.style\
.set_caption("Weather Conditions")\
.background_gradient(axis=None, vmin=1, vmax=5, cmap='BuPu',subset=['Tokyo'])\
.format(rain_condition)\
.format_index(change_index)
```

```python
import numpy as np
import pandas as pd

scores = np.random.randint(50, 101, size=(50, 5))
df = pd.DataFrame(scores,
             columns=['國文', '英文', '數學', '地理', '歷史'],
             index=range(1,51)
             )
df.index.name = '學號'
df.columns.name = '科目'
df
```

```python
def less_60(v):
    if v < 60:
        return 'background-color:red'
df.style.applymap(less_60)
```

```python
def less_cells(row):
    if len(row[row < 60]) >= 3:  # type: ignore
        return ['background-color:red'] * len(row)
    else:
        return [''] * len(row)

df.style.apply(less_cells, axis=1)
```
